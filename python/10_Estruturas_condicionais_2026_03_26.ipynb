## 04 – Escreva o programa em Python:

```python
import numpy as np

array_random = np.random.normal(loc=10, scale=3, size=5)
valores_superiores = array_random[array_random > 10]
quantidade = len(valores_superiores)
print(f"Valores superiores a 10: {valores_superiores}")
print(f"Array gerado: {array_random}")
print(f"Quantidade de valores: {quantidade}")

if quantidade > 0:
  print(f"Existem {quantidade} valores superiores a 10")
else:
  print("Nenhum valor é superior a 10")
```

```python
import pandas as pd
genes = ["BRCA1", "TP53", "APOE", "MTHFR", "EGFR"]
dados_gene = {
    'BRCA1': [5, 'BR', 1, False],
    'TP53': [4, 'TP', 0, False],
    'APOE': [4, 'AP', 1, False],
    'MTHFR': [5, 'MT', 0, False],
    'EGFR': [4, 'EG', 0, False]
}
df_genes = pd.DataFrame.from_dict(dados_gene, orient='index')
df_genes.columns = ['Tamanho', 'Prefixo', 'Contagem_A', 'Possui_HU']
print(df_genes)
```

```python
import pandas as pd

df_genes.to_csv ('dados_geneticos.csv')
df_importado = pd.read_csv('dados_geneticos.csv', index_col=0)
df_importado.columns = ['Tamanho_nucleotideo', 'Sigla_Prefixo', 'Contagem_A', 'Status_HU']

primeiro_gene = df_importado.index[0]
num_caracteres = len(primeiro_gene)

print(f"O primeiro gene é: {primeiro_gene}")
print(f"O tamanho do primeiro gene é: {num_caracteres}")

if 2 <= num_caracteres <= 4:
  print("Informações detalhadas do gene")
  print(df_importado.loc[primeiro_gene])

elif num_caracteres < 2:
  print("Informações gerais do gene")

  print(df_importado.loc[df_importado['Tamanho_nucleotideo'] > 4])

else:
  print("O nome possui mais de 4 caracteres. Listando apenas quem é superior")
  superiores = df_importado[df_importado['Tamanho_nucleotideo'] > 4]
  print(superiores)
```
