## 01 – Dados as operações abaixo, qual o resultado da operação lógica:
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2
resultado_logico = p and q and r and s and t
print(resultado_logico)
```

```python
Conforme visto em aula, o operador lógico not poderia ser usado para inverter o resultado, ja que ele inverte o valor booleano. Se usar o 'or' também teriamos uma inversão de resultado, ja que o or exige que apenas um seja verdadeiro.
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
<br>u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] > ['a',True,1]
t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

resultado_final = p or q or r or s or t or u
print(resultado_final)

resultado_invertido = not (p or q or r or s or t or u)
print(resultado_invertido)
```

## 03 – No exercício 3 da aula 8 foi importado um arquivo .csv para um objeto da classe DataFrame pandas. Importe aqui o mesmo objeto e arredonde e edite o index como realizado no exercício 3 da aula 8. Usando operadores responda, a segunda linha é igual a terceira ou a terceira linha é igual a quarta?

```python
import pandas as pd
import numpy as np

dados = np.array([
    [2.6, 2.2, 34.46, 0.25, 9.878],
    [0.5, 1.61, 1.48, 5.87, 2.365],
    [0.5, 0.7, 2.6, 2.2, 1.5],
    [34.46, 0.25, 0.5, 1.61, 9.205],
    [1.48, 5.87, 0.5, 0.7, 2.138],
    [7.908, 2.126, 7.908, 2.126, 5.017]
])

colunas = ['Col_1', 'Col_2', 'Col_3', 'Col_4', 'Medias_Linhas']
indices = ['Linha_1', 'Linha_2', 'Linha_3', 'Linha_4', 'Linha_5', 'Medias_Colunas']
df_original = pd.DataFrame(dados, index=indices, columns=colunas)

df_original.to_csv('exercicio_10.csv')
```

```python

df_novo = pd.read_csv ('exercicio_10.csv', index_col = 'Unnamed: 0')
df_novo = df_novo.round(3)

comparacao_1 = (df_novo.iloc[1] == df_novo.iloc[2]).all()
comparacao_2 = (df_novo.iloc[2] == df_novo.iloc[3]).all()

resultado_final = comparacao_1 or comparacao_2
print(resultado_final)
```

## 04 – Usando operadores verifique se todos os valores contidos na primeira linha são maiores ou iguais a 10.

```python
resultado_04 = (df_novo.iloc[0] >= 10).all()
print(resultado_04)
```

## 05 – Usando operadores verifique se todos os valores contidos na terceira coluna estão entre 0 inclusive e 40.

```python
coluna_3 = df_novo.iloc[:, 2]
resultado_05 = ((coluna_3 >= 0) & (coluna_3 <= 40)).all()
print(resultado_05)
```

## 06 – Usando operadores verifique se algum dos valores contidos na última linha está acima de 10.

```python
resultado_06 = (df_novo.iloc[-1] > 10).any()
print(resultado_06)
```

## 07 – Recupere a DataFrame contendo apenas valores que estejam acima de 5 e abaixo de 7.

```python
df_filtrado = df_novo[(df_novo > 5) & (df_novo < 7)]
print(df_filtrado)
```

## 08 – Recupere a DataFrame contendo apenas valores que estejam abaixo de 5 inclusive ou acima de 7.

```python
df_filtrado_ou = df_novo[(df_novo <= 5) | (df_novo >= 7)]
print(df_filtrado_ou)
```

## 09 – Converta as duas últimas colunas do objeto DataFrame para dois objetos da classe numpy array. Algum dos valores contidos no primeiro array estão entre 2 inclusive e 5 inclusive?

```python
array_col4 = df_novo.iloc[:,-2].values
array_medias = df_novo.iloc[:,-1].values
resultado_09 = ((array_col4 >= 2) & (array_col4 <= 5)).any()
print(resultado_09)
```

## 10 – Os valores contidos nos dois arrays são superiores à -1 e inferiores à 12?

```python
condicao_array1 = ((array_col4 > -1) & (array_col4 < 12)).all()
condicao_array2 = ((array_medias > -1) & (array_medias < 12)).all()
resultado_10 = condicao_array1 and condicao_array2
print(resultado_10)
```
