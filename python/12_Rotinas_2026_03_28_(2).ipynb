Instancie um objeto da classe NumPy array, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.<br><br>
Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

R.: Primeiro usei da biblioteca NumPy para criar um array com distribuição normal e ter o números aleatorios necessários. Com essa função é possível alterar o DP e média, já a filtragem dos dados foi realizada com indexação booleana facilitando a contagem final com len().

## 04 – Escreva o programa em Python:

```python
import numpy as np

def analisar_valores_aleatorios (media=10, desvio=3, quantidade=5):
    valores = np.random.normal(media,desvio,quantidade)
    superiores = valores[valores > media]
    quantidade_sup = len(superiores)
    return valores, superiores, quantidade_sup

todos, acima_media, total = analisar_valores_aleatorios()

print(f"valores gerados: {todos}", "valores superiores a 10: {acima_media}", "quantidade de valores superiores a 10: {total}")
```

## Desenvolva um programa de computador que retorne a média de um conjunto de valores.

## 06 – Descreva o racional da sua resposta:

R.: Foquei na modularização do cálculo. A função que utilizei foi projetada para receber um objeto iterável. O programa então utiliza a função sum() para a soma total e len() para determinar o tamanho do conjunto. A média é obtida pela divisão desses dois valores.

## 07 – Escreva o programa em Python:

```python
def calcular_media(valores):
  if len(valores) == 0:
    return 0
  soma = sum(valores)
  quantidade = len(valores)
  media = soma / quantidade

  return media

Conjunto_de_valores = [1, 3, 5, 7, 9, 10]
resultado = calcular_media(Conjunto_de_valores)
print(f"o conjunto de valores é: {Conjunto_de_valores}")
print(f"A média dos valores é: {resultado}")
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def divisao_de_dois_objetos(a, b):
  print(f"Objeto 1: Classe {type(a)} | Valor: {a}")
  print(f"Objeto 2: Classe {type(b)} | Valor: {b}")
  if b == 0:
    return "Não é possível dividir por zero"
  resultado = a / b
  return resultado

valor1 = 13
valor2 = 3
resultado = divisao_de_dois_objetos(valor1, valor2)
print(f"O resultado da divisão é: {resultado}")


```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
import numpy as np
def divisao_composta(obj1, obj2):
  print(f"Objeto 1: Classe {type(obj1)} | Valor: {obj1}")
  print(f"Objeto 2: Classe {type(obj2)} | Valor: {obj2}")

  arr1 = np.array(obj1)
  arr2 = np.array(obj2)

  if len(arr1) != len(arr2):
    return "Erro: os arrays devem ter o mesmo tamanho"

  if 0 in arr2:
    return "Erro: o segundo objeto tem valor zero"
  resultado = arr1 / arr2
  return resultado

lista_a = [100,200,300]
lista_b = [3,5,10]

final = divisao_composta(lista_a, lista_b)
print(f"resultado da divisão composta: {final}")
```
