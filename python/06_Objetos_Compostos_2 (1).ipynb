## 5 – Verifique se todos os nomes dos genes tem ou não os caracteres 'hu'.

```python
genes = ["BRCA1", "TP53", "APOE", "MTHFR", "EGFR"]
verificacao = []
for gene in genes:
    if 'hu' in gene:
        verificacao.append(True)
    else:
        verificacao.append(False)
print(verificacao)
```

## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
import pandas as pd
df_seguro = pd.read_csv('/content/data.csv')
classes_colunas = df_seguro.dtypes.astype(str).tolist()
contagem_classes = df_seguro.dtypes.value_counts().to_dict()
colunas_alvo = df_seguro.iloc[:, [0,3,8]]
estatisticas = colunas_alvo.describe()
resultado_seguro= {
    'numero_entradas': len(df_seguro),
    'classes_colunas': classes_colunas,
    'contagem_por_classes': list (contagem_classes.values()),
    'medias': estatisticas.loc['mean'].tolist(),
    'desvio_padrao': estatisticas.loc['std'].tolist(),
    'terceiro_quartil': estatisticas.loc['75%'].tolist()
}
print(resultado_seguro)
```
