# Exercício 02:

Em Python, o método com nome count() faz parte do objeto da classe string. Este objeto possuí outros atributos e métodos. Qual o código que pode ser usado para verificar se count() é um método também presente no objeto da classe integer?

```python
#Para fazer essa verificação vou utilizar a função dir (). ELa vai me retornar uma lista com todos os atributos para aquele dado. Então o codigo que eu usaria para verificar isso seria:

possui_count = 'count' in dir(int)

print("O tipo 'int' possui o método 'count'? {possui_count}")
```
