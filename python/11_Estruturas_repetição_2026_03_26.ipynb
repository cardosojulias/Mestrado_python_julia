## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0.
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: Para gerar a lista com os 5000 números inteiros, comecei usando a função range combinada com list. Utilizei a estrutura for para percorrer cada item da lista. Por fim, usei o operador de módulo % para calcular por divisão os pares e impares.

## 03 – Escreva o programa em Python:

```python
lista_valores = list (range(5000))

pares = 0
impares = 0

for numero in lista_valores:
  if numero % 2 == 0:
    pares +=1
  else:
    impares +=1

print(f'Quantidade de pares: {pares}')
print(f'Quantidade de impares: {impares}')
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
lista_valores = list (range(5000))
list_pares = [n for n in lista_valores if n % 2 == 0]
list_impares = [n for n in lista_valores if n % 2 != 0]

total_pares = len(list_pares)
total_impares = len(list_impares)

print(f'Total de Quantidade de pares: {total_pares}')
print(f'Total de Quantidade de impares: {total_impares}')
```

## Importe o .csv exportado na última aula e instancie um objeto da classe DataFrame.

Dos números de letras nos genes, quais são pares e quantos são impares?

```python
dados_gene = {
    'BRCA1': [5, 'BR', 1, False],
    'TP53': [4, 'TP', 0, False],
    'APOE': [4, 'AP', 1, False],
    'MTHFR': [5, 'MT', 0, False],
    'EGFR': [4, 'EG', 0, False]
}

import pandas as pd

df_genes = pd.DataFrame.from_dict(dados_gene, orient='index')
df_genes.columns = ['Tamanho_nucleotideo', 'Sigla_Prefixo', 'Contagem_A', 'Status_HU']
df_genes.to_csv('dados_geneticos.csv')

print("Arquivo 'dados_geneticos.csv' criado com sucesso neste notebook!")
```

## 05 – Descreva o racional da sua resposta:

R.: Para esse código transformamos os dados qualitativos em quantitativos, para isso, apos carregar o arquivo em CSV, utilizei de iteração para medir o comprimento de cada nome de gene. Utilizei do operador de módulo % dentro de um laço de repetição para contar ao dividir por 2 quais seriam impar e quais seriam par, permitindo classificar os nomes rapidamente.

## 07 – Escreva o programa em Python:

```python
import pandas as pd

df_genes = pd.read_csv('dados_geneticos.csv', index_col=0)
tamanhos = [len(gene) for gene in df_genes.index]

pares = 0
impares = 0

for n in tamanhos:
  if n % 2 == 0:
    pares+= 1
  else:
    impares +=1

print(f'Quantidade de genes com número de letras par: {pares}')
print(f'Quantidade de genes com número de letras impar: {impares}')
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
import pandas as pd

df_genes = pd.read_csv('dados_geneticos.csv', index_col=0)

pares_lista = [len(gene) for gene in df_genes.index if len(gene) % 2 ==0]
impares_lista = [len(gene) for gene in df_genes.index if len(gene) % 2 !=0]

total_pares = len(pares_lista)
total_impares = len(impares_lista)

print(f"total de genes com letras pares: {total_pares}")
print(f"total de genes com letras impares: {total_impares}")
```

## 09 – Usando a coluna de presença de hu no nome do gene, itere os index deste objeto usando a função enumerate(). Estes index podem ser usados para imprimir cada um dos valores presentes na coluna dos dois primeiros caracteres dos nomes dos genes?

```python
Sim, podem ser utilizados para imprimir cada um dos valores presentes na coluna dos dois primeiros caracteres dos nomes dos genes.
```

```python
import pandas as pd

df_genes = pd.read_csv('dados_geneticos.csv', index_col=0)

for i, gene in enumerate(df_genes.index):
    print(f"Índice: {i}, Nome do Gene: {gene}")
prefixo = df_genes['Sigla_Prefixo'].iloc[i]
print(f"Posição {i} | Gene: {gene} | Dois primeiros caracteres: {prefixo}")
```

## 10 – Selecione qualquer uma das instâncias contidas na DataFrame. É possível recuperar as chaves e os valores com ao menos 3 estruturas de repetição diferentes?

```python
Sim, é possível recuperar as chaves e os valores com estruturas de repetição diferentes.
```

```python
import pandas as pd

df_genes = pd.read_csv('dados_geneticos.csv', index_col=0)
instancia = df_genes.iloc[0]

print(f"analisando o gene: {instancia.name}")
print("--- Estrutura 1: For simples ---")
for coluna in instancia.index:
    valor = instancia[coluna]
    print(f"Chave: {coluna} | Valor: {valor}")

print("\n--- Estrutura 2: For com .items() ---")
for chave, valor in instancia.items():
    print(f"Chave: {chave} | Valor: {valor}")

print("\n--- Estrutura 3: While com contador ---")
i = 0
while i < len(instancia):
    chave = instancia.index[i]
    valor = instancia.iloc[i]
    print(f"Chave: {chave} | Valor: {valor}")
    i += 1
```
